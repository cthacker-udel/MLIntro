```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_digits
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
digits = load_digits()
```

```python
digits_df = pd.DataFrame(digits.data, columns=digits.feature_names)
digits_df.head()

digits_df['number'] = digits.target

digits_df.head()
```

Using GridSearchCV to find the optimal parameters with not as tight of a regard about runtime

```python
train_x, test_x, train_y, test_y = train_test_split(digits_df.drop(columns=['number']), digits_df['number'])
```

```python
train_x.head()
```

```python
parameters = {
    'svm': {
        'model': svm.SVC(gamma='auto'),
        'params': {
            'C': [1, 10, 20, 30, 50],
            'kernel': ['rbf', 'linear']
        }
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators': [1, 5, 6, 10]
        }
    },
    'logistic_regression': {
        'model': LogisticRegression(),
        'params': {
            'C': [1, 5, 10, 20]
        }
    }
}
```

```python
total_scores = []

for model_name in parameters:
    clf = GridSearchCV(parameters[model_name]['model'], parameters[model_name]['params'], cv=5, return_train_score=False)
    clf.fit(train_x, train_y)
    total_scores.append({
        'model': model_name,
        'best_score': 'clf.best_score_',
        'best_params': clf.best_params_
    })

total_scores
```
