Finding all NaN columns within the X data

```python
nan_columns = [col for col in X.columns if X[col].hasnans]

nan_columns
```

We want to approach the nan columns accordingly, measure the percentage of how many rows have NaN values before we start removing or imputing columns

```python
def get_nan_percentage(column_name, dataset, total_rows):
    """
    Gets a %age of the NaN values within the column

    Args:
        column_name (_type_): The name of the column to analyze
        dataset (_type_): The dataset we are analyzing
        total_rows (_type_): The total # of rows the dataset has (used to calculate percentage)

    Returns:
        _type_: The percentage
    """
    nan_count = len(dataset[dataset[column_name].isna()])
    return round((nan_count / total_rows) * 100, 1)
```

```python
percentages = {}
total_dataset_rows = X.shape[0]
for each_column in nan_columns:
    percentages[each_column] = '{}%'.format(get_nan_percentage(each_column, X, total_dataset_rows))

percentage_items = sorted(list(percentages.items()), key=lambda x: float(x[1].replace('%', '')))

pprint(percentage_items)

col_types = []

for each_item in percentage_items:
    col_types.append('{} = {}'.format(each_item[0], X[each_item[0]].dtype))
```

Judging off this printout, there are columns that have barely any significance within the dataset, such as Alley, which has a 93.8% NaN count, and also PoolQC, which has a 99.5% rating, those ones can be dropped with little fallout.

There are 3 approaches to dealing with NaN columns: 1) Drop NaN columns, 2) Impute NaN columns with a constant or statistical method, 3) Extend imputation and perform a variant of One-Hot encoding, where we add another column that signifies the respective row was imputed, with either a 1 or 0.

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer, SimpleImputer
from sklearn.ensemble import RandomForestRegressor

nan_modified_x = X.copy()
nan_modified_x_val = X_val.copy()

#2 impute with statistical number

nan_modified_x['LotFrontage'].fillna(nan_modified_x['LotFrontage'].mean(), inplace=True)
nan_modified_x['MasVnrArea'].fillna(nan_modified_x['MasVnrArea'].mean(), inplace=True)
nan_modified_x_val['LotFrontage'].fillna(nan_modified_x_val['LotFrontage'].mean(), inplace=True)
nan_modified_x_val['MasVnrArea'].fillna(nan_modified_x_val['MasVnrArea'].mean(), inplace=True)

#1 drop NaN columns
nan_modified_x.drop(columns=['MiscFeature', 'PoolQC', 'Alley', 'Fence', 'FireplaceQu'], inplace=True)
nan_modified_x_val.drop(columns=['MiscFeature', 'PoolQC', 'Alley', 'Fence', 'FireplaceQu'], inplace=True)

# access the NaN columns left after dropping NaN columns
modified_nan_columns = [col for col in nan_modified_x if nan_modified_x[col].hasnans]
modified_nan_validation_columns = [col for col in nan_modified_x_val if nan_modified_x_val[col].hasnans]

# create simple imputer imputing by most frequent algorithm
imputer = SimpleImputer(strategy='most_frequent')

# Access imputed columns from imputing NaN columns
imputed_cols = pd.DataFrame(imputer.fit_transform(nan_modified_x[modified_nan_columns]))
validation_imputed_cols = pd.DataFrame(imputer.fit_transform(nan_modified_x_val[modified_nan_validation_columns]))

# Copying the index over
imputed_cols.index = nan_modified_x[modified_nan_columns].index
validation_imputed_cols.index = nan_modified_x_val[modified_nan_validation_columns].index

# Copying over the columns from the original to the imputed df
imputed_cols.columns = modified_nan_columns
validation_imputed_cols.columns = modified_nan_validation_columns

# Copying over the imputed column data to the original df
nan_modified_x[modified_nan_columns] = imputed_cols
nan_modified_x_val[modified_nan_validation_columns] = validation_imputed_cols

X = nan_modified_x.copy()
X_val = nan_modified_x_val.copy()

(X.shape, X_val.shape)
```

Now that we've fixed all NaN values in the dataset, we can extract the target variable from the X dataset and the X_validation dataset

```python
y = X['SalePrice'] # Extract the target variable from the dataset

X.drop(columns=['SalePrice'], inplace=True) # remove the column from the dataset
```

Now we attempt to verify the performance of all the categorical variable approaches, and how they affect the score of the model

```python
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor

def get_categorical_mae(x_train, x_test, y_train, y_test):
    """
    Gets the MAE of the random forest regressor

    Args:
        x_train (_type_): The features
        y_train (_type_): Target column
        x_test (_type_): Test feature dataset
        y_test (_type_): Test target column

    Returns:
        _type_: _description_
    """
    tree = RandomForestRegressor(random_state=0)
    tree.fit(x_train, y_train)
    predictions = tree.predict(x_test)
    return mean_absolute_error(predictions, y_test)
```

Approach 1: Drop all categorical columns without surgical decisions (basically meaning that we just drop them regardless of whether they are prominent in the dataset)

```python
drop_c_X = X.copy() # copied X dataset
drop_c_X_validation = X_val.copy() # copy the validation dataset
```

```python
categorical_columns = [x for x in drop_c_X.columns if drop_c_X[x].dtype == 'object']

drop_c_X.drop(columns=categorical_columns, inplace=True)
```

```python
## split the data

cdrop_train_x, cdrop_val_x, cdrop_train_y, cdrop_val_y = train_test_split(drop_c_X, y, train_size=0.8, test_size=0.2, random_state=0)

get_categorical_mae(cdrop_train_x, cdrop_val_x, cdrop_train_y, cdrop_val_y)
```

Approach 2: Ordinal encoding

```python
from sklearn.preprocessing import OrdinalEncoder

oe_x_df = X.copy()
oe_x_val_df = X_val.copy()
ordinal_encoder = OrdinalEncoder()

# Grabbing all categorical columns, which contain an column type of "object"
oe_categorical_columns = [col for col in oe_x_df.columns if oe_x_df[col].dtype == 'object']

# Create a new dataframe of the ordinal encoding
oe_encoded_categorical_columns = pd.DataFrame(ordinal_encoder.fit_transform(oe_x_df[oe_categorical_columns]))
oe_encoded_categorical_columns.index = oe_x_df.index # copy index over
oe_encoded_categorical_columns.columns = oe_categorical_columns # copy columns over
oe_x_df[oe_categorical_columns] = oe_encoded_categorical_columns # apply ordinal encoding to dataframe

# repeat same process with validation data
oe_encoded_val_categorical_columns = pd.DataFrame(ordinal_encoder.fit_transform(oe_x_val_df[oe_categorical_columns]))
oe_encoded_val_categorical_columns.index = oe_x_val_df.index
oe_encoded_val_categorical_columns.columns = oe_categorical_columns
oe_x_val_df[oe_categorical_columns] = oe_encoded_val_categorical_columns
```

```python
oe_x_train, oe_x_val, oe_y_train, oe_y_val = train_test_split(oe_x_df, y) # splitting the encoded dataframe
```

```python
get_categorical_mae(oe_x_train, oe_x_val, oe_y_train, oe_y_val)
```

Approach 3: One-Hot Encoding
