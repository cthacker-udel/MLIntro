```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn

from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data[0]
```

```python
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(digits.data, digits.target, test_size=0.1)

train_x
```

```python
from sklearn.linear_model import LogisticRegression

digits_model = LogisticRegression()
```

```python
digits_model.fit(train_x, train_y)
```

```python
digits_model.score(test_x, test_y)
```

```python
plt.matshow(digits.images[67])
```

```python
digits.target[67]
```

```python
digits_model.predict([digits.data[67]])
```

```python
digits_model.predict(digits.data[0:5])
```

```python
y_predicted = digits_model.predict(test_x)

from sklearn.metrics import confusion_matrix

confusion_matrix = confusion_matrix(test_y, y_predicted)

%matplotlib inline
plt.figure(figsize=(10, 7))
sn.heatmap(confusion_matrix, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
