```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
import pandas as pd
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

### Loading in the dataset

```python
titanic_data = pd.read_csv('titanic.csv')

titanic_data.head()
```

```python
# encode the labels

class_col = titanic_data['Pclass']

class_col.head()
```

```python
sex_encode = LabelEncoder()

sex_col = sex_encode.fit_transform(titanic_data['Sex'])

titanic_data['Sex'] = sex_col

titanic_data['Age'].fillna(titanic_data['Age'].mean(), inplace=True)

titanic_data.head()
```

```python
input_x = titanic_data.drop(columns=['PassengerId', 'Survived', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'])

input_y = titanic_data['Survived']

train_x, test_x, train_y, test_y = train_test_split(input_x, input_y, test_size=0.2)

train_x.head()
```

```python
titanic_tree = tree.DecisionTreeClassifier()

titanic_tree.fit(train_x, train_y)
```

```python
titanic_tree.score(test_x, test_y)
```
